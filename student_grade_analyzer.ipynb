OBJECTIVE:

Student Gradebook Analyzer is a Python project developed using the Pandas library. It reads student data from a CSV file and performs various statistical analyses such as Average, Median, Maximum, Minimum, Grade Calculation, and Student Ranking.

This project demonstrates basic data analysis using Python and Pandas.

```python
import pandas as pd 
df=pd.read_csv("STUDENTS.csv")
print(df.shape)
print(df.dtypes)
print(df.info())
print(df.isnull().sum())
```

```python
import pandas as pd
df=pd.read_csv("STUDENTS.csv")
df["Average"]=df[["Maths","Physics","Chemistry","English"]].mean(axis=1)
df["Median"]=df[["Maths","Physics","Chemistry","English"]].median(axis=1)
df["Maximum"]=df[["Maths","Physics","Chemistry","English"]].max(axis=1)
df["Minimum"]=df[["Maths","Physics","Chemistry","English"]].min(axis=1)
def grade(avg):
    if avg>=90 and avg<100:
        return "A+"
    elif avg>=80 and avg<90:
        return "A"
    elif avg>=70 and avg<80:
        return "B+"
    elif avg>=60 and avg<70:
        return "B"
    elif avg>=50 and avg<60:
        return "C"    
    elif avg>=40 and avg<50:
        return "D"    
    else:
        return "FAIL"
df["Grade"]=df["Average"].apply(grade)   
sort_by_avg=df.sort_values(by="Average",ascending=False)
sort_by_avg.head()
```

CONCLUSION:

This project successfully analyzes student performance using Python and Pandas. It calculates Average, Median, Maximum, Minimum, assigns Grades, and sorts students based on their Average marks. The project demonstrates how data can be processed efficiently and accurately, making student performance analysis faster and more reliable.
